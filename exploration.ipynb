```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
column_list_data = ["uid", "iid", "rating", "tsp"]

pd_data = pd.read_csv('ml-100k/u.data', delimiter='\t', names=column_list_data) 
pd_data.head()
```

```python
column_list_item = ["iid", "title", "release", "vrelease", "url", "unknown", 
                    "action", "adventure", "animation", "childrens", "comedy",
                   "crime", "documentary", "drama", "fantasy", "noir", "horror",
                   "musical", "mystery", "romance", "scifi", "thriller",
                   "war", "western"]

pd_item = pd.read_csv('ml-100k/u.item', delimiter='|', names=column_list_item, encoding = "ISO-8859-1") 
pd_item.head()
```

```python
pd_item["title"] = pd_item["title"].apply(lambda x: ' '.join(x.split()[:-1]))
```

```python
pd_item.head()
```

```python
column_list_user = ["uid", "age", "gender", "occupation", "zip"]

pd_user = pd.read_csv('ml-100k/u.user', delimiter='|', names=column_list_user, encoding = "ISO-8859-1") 
pd_user.head()
```

```python
ratings = np.zeros((pd_user.shape[0]+1, pd_item.shape[0]+1))
ratings = ratings - 1

np_data = pd_data.to_numpy()

for i in range(np_data.shape[0]) :
    ratings[np_data[i,0], np_data[i,1]] = np_data[i,2]
```

```python
present = (ratings[1,:] > 0)
print(ratings[1,:])
print(ratings[1,:]*present)
```

```python
columns_imdb_titles = ["imdb_mid", "title", "imdb_idx",
    "kind", "year", "imdb_id", "phonetic", "episode_id",
    "season", "episode", "series_years", "md5"]

pd_titles = pd.read_csv('ml-100k/title.csv', delimiter=',', names=columns_imdb_titles, encoding = "ISO-8859-1") 
pd_titles.head()
```

```python
pd_movie_ext = pd_item.merge(pd_titles, how="left", on="title")
pd_movie_ext.head()
```

```python
import math
```

```python
pd_movie_unmerged = pd_movie_ext[pd_movie_ext.apply(lambda x: math.isnan(x["imdb_mid"]), axis=1)]
```

```python
pd_movie_unmerged
```

```python
articles = [" A", " The"]

pd_movie_merged2 = pd_movie_unmerged[pd_movie_unmerged.apply(lambda x : x["title"].split(",")[-1] in articles, axis=1)]
```

```python
pd_movie_merged2["title"] = pd_movie_merged2["title"].apply(lambda x : x.split(",")[-1][1:] + " " + ','.join(x.split(",")[:-1]))
pd_movie_merged2
```
